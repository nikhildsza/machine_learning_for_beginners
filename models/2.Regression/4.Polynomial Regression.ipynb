Polynomial Regression

Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv("/home/nikhilds/GitHub/machine_learning_for_beginners/datasets/1.Regression/PolynomialLinearRegression.csv")
X = dataset.iloc[:,1:-1].values # We want only the values of the levels
Y = dataset.iloc[:,-1].values
```

Training the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

Training the Polynomial Linear Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,Y)
```

Visualizing the Linear Regression Results

```python
plt.scatter(X,Y,color = 'red')
plt.plot(X,lin_reg.predict(X),color = 'blue')
plt.title("Linear Regression Results")
plt.xlabel("Positon Level")
plt.ylabel("Salaries")
plt.show()
```

Visualizing the Polynomial Regression Results

```python
plt.scatter(X,Y,color = 'red')
plt.plot(X,lin_reg2.predict(X_poly),color = 'blue')
plt.title("Polynomial Regression Results")
plt.xlabel("Positon Level")
plt.ylabel("Salaries")
plt.show()
```

Visualizing for better results

```python
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,Y,color = 'red')
plt.plot(X_grid,lin_reg2.predict(poly.fit_transform(X_grid)),color='Blue')
plt.title("Polynomial Regression Results")
plt.xlabel("Positon Level")
plt.ylabel("Salaries")
plt.show()
```

Predicting the Linear Regression Results

```python
lin_reg.predict([[6.5]])
```

Predicting the Polynomial Regression Results

```python
lin_reg2.predict(poly.fit_transform([[6.5]]))
```
