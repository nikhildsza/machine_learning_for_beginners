Hierarchical Clustering

Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv("/home/nikhilds/GitHub(ML)/machine_learning_for_beginners/datasets/3.Clustering/HierarchicalClustering.csv")
X = dataset.iloc[:,[3,4]].values
```

Using dendrograms to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrograms = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Eucledian Distance")
plt.show()
```

Training the Hierarchical Clustering Model

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5,  metric='euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

Visualizing the results

```python
plt.scatter(X[y_hc == 0,0],X[y_hc == 0,1],s=100,c='red',label="Cluster1")
plt.scatter(X[y_hc == 1,0],X[y_hc == 1,1],s=100,c='blue',label="Cluster2")
plt.scatter(X[y_hc == 2,0],X[y_hc == 2,1],s=100,c='yellow',label="Cluster3")
plt.scatter(X[y_hc == 3,0],X[y_hc == 3,1],s=100,c='green',label="Cluster4")
plt.scatter(X[y_hc == 4,0],X[y_hc == 4,1],s=100,c='pink',label='Cluster5')
plt.title("Hierarchical Clustering")
plt.xlabel("Salary Scope")
plt.ylabel("Age")
plt.legend()
plt.show()
```
